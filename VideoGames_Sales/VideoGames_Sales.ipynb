***3.4. Menganalisis korelasi critic score dan user score terhadap penjualan game***

Sekarang kita Lihat bagaimana ulasan pengguna dan para profesional memengaruhi penjualan pada platform populer (PS4 dan X360).

```python
new_df_games.head()
```

```python
# Mengambil data salah satu platform yang populer (PS4)
ps4_score = new_df_games.groupby(['platform', 'name'])['total_sales', 'critic_score', 'user_score'].sum().query('platform == "PS4" & critic_score > 0 & user_score > 0').reset_index()
    
ps4_score
```

```python
ps4_score.corr()['total_sales']
```

    - Nilai korelasi antara critic_score dengan total sales PS4 adalah 0.405. Artinya critic score bisa meningkatkan total 
    sales PS4, namun tidak terlalu kuat.
    - Sedangkan korelasi antara user score dengan total sales PS4 adalah -0.03 (mendekati 0). Artinya user score tidak
    mempengaruhi total sales PS4.

```python
# Scatterplot korelasi antara critic score dengan total penjualan PS4
plt.figure(figsize=(15,5))
sns.scatterplot(x='critic_score', y='total_sales', data=ps4_score)
plt.title('Correlation between Critic Score and Total Sales PS4')
```

Scatter plot antara critic score dengan total sales PS4 di atas memiliki pola cenderung naik, artinya semakin tinggi nilai critic score, total sales PS4 ikut naik.

# Tahap 5. Melakukan Uji Hipotesis

***1. Hipotesis rata-rata rating pengguna (user_score) platform Xbox One dan PC adalah sama***

    - H0 =  Rata-rata rating pengguna (user_score) platform Xbox One dan PC adalah SAMA
    - H1 =  Rata-rata rating pengguna (user_score) platform Xbox One dan PC adalah BERBEDA

```python
# Library uji statistik
from scipy.stats import ttest_1samp, shapiro, mannwhitneyu
```

```python
# Menyiapkan dataset untuk platform XOne
user_xone = new_df_games.loc[(new_df_games['platform']=='XOne') & (new_df_games['user_score'] > 0), 'user_score']
sns.histplot(user_xone)
plt.show()
```

```python
# Menyiapkan dataset untuk platform PC
user_pc = new_df_games.loc[(new_df_games['platform']=='PC') & (new_df_games['user_score'] > 0), 'user_score']
sns.histplot(user_pc)
plt.show()
```

```python
# Membuat function uji kenormalan
def normality_test(series):
    p_value_shapiro = shapiro(series).pvalue
    print(f'P-Value: {p_value_shapiro}')
    print(f'Alpha  : {alpha}')
    if p_value_shapiro >= alpha:
        print('H0 Diterima: Distribusi data tersebut adalah normal')
    else:
        print('H1 Diterima: Distribusi data tersebut adalah tidak normal')
```

```python
# Uji Kenormalan dari distribusi user_score XOne dan PC
alpha = 0.05
normality_test(user_xone)
print('---------------------------')
normality_test(user_pc)
```

Distribusi data Xone dan PC BUKAN distribusi NORMAL

```python
# Uji hipotesisnya
# Karena distribusi datanya bukan distribusi normal maka uji hipotesisnya menggunakan metode Mann-Whitney U Test
alpha = 0.05
p_value = mannwhitneyu(user_xone, user_pc, nan_policy='omit').pvalue
print(f'Alpha  : {alpha}')
print(f'P-Value: {p_value}')
if p_value >= alpha:
    print('H0 Diterima: Rata-rata rating pengguna (user_score) platform Xbox One dan PC adalah SAMA')
else:
    print('H1 Diterima: Rata-rata rating pengguna (user_score) platform Xbox One dan PC adalah BERBEDA')
```

Hipotesis H0 diterima bahwa rata-rata rating pengguna (user score) platform Xbox One dan PC adalah sama.
