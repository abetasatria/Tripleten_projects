***D. Frekuensi pengguna kembali untuk menggunakan produk***

```python
first_visit = visits.groupby('uid').agg({'session_start_ts': 'min'}).reset_index()
first_visit.columns = ['uid', 'first_time_use']
first_visit.head()
```

```python
first_visit['first_session_date'] = first_visit['first_time_use'].dt.date
first_visit['first_session_week'] = first_visit['first_time_use'].astype('datetime64[W]')
first_visit['first_session_month'] = first_visit['first_time_use'].astype('datetime64[M]')
first_visit['first_session_year'] = first_visit['first_time_use'].astype('datetime64[Y]')
```

```python
first_visit
```

```python
#Menggabungkan tabel first_visit dengan tabel visit
new_visits = pd.merge(visits, first_visit, on= 'uid')
new_visits.head()
```

```python
#Menghitung jarak hari dan bulan antara sesi pertama dengan sesi berikutnya
new_visits['age_days'] = (new_visits['session_date'] - new_visits['first_session_date']).dt.days
new_visits['age_month'] = round(new_visits['age_days'] / 30,0).astype('int')
new_visits.sample(n=8)
```

```python
#Menghitung jumlah users per kohort 
cohorts = new_visits.pivot_table(index='first_session_month', columns='age_month', values='uid', aggfunc='nunique')
cohorts.fillna('')
```

```python
#Menghitung users yang tetap aktif (retention)
retention = pd.DataFrame()
for col in cohorts.columns:
    retention = pd.concat([retention, cohorts[col]/cohorts[0]], axis=1)
retention.columns = cohorts.columns
retention.fillna("")
```

***D. Jumlah uang yang mereka kontribusikan (LTV)***

```python
first_buy.head()
```

```python
cohort_sizes = first_buy.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.rename(columns={'uid': 'n_buyers'}, inplace=True)
cohort_sizes
```

```python
cohorts = pd.merge(orders, first_buy, on='uid')
cohorts_agg = cohorts.groupby(['first_order_month', 'order_month']).agg({'revenue': 'sum'}).reset_index()
cohorts_agg.head()
```

```python
#Menambahkan age_month
cohorts_agg['age_month'] = (cohorts_agg['order_month'] - cohorts_agg['first_order_month']).dt.days
cohorts_agg['age_month'] = round(cohorts_agg['age_month']/30,0).astype('int')
cohorts_agg.columns = ['first_order_month', 'order_month', 'revenue', 'age_month']
cohorts_agg
```

```python
#Menghitung gross profit dan ltv
margin_rate = 0.5

cohorts_report = pd.merge(cohort_sizes, cohorts_agg, on='first_order_month')
cohorts_report['gp'] = cohorts_report['revenue'] * margin_rate
cohorts_report['ltv'] = cohorts_report['gp'] / cohorts_report['n_buyers']
cohorts_report
```

```python
#Menghitung LTV per bulan tiap kohort
cohorts_ltv = cohorts_report.pivot_table(
    index='first_order_month',
    columns='age_month', 
    values='ltv',
    aggfunc='mean').cumsum(axis=1)
cohorts_ltv.round(2).fillna('')
```

Secara rata-rata, setiap pelanggan dari kohort pertama berkontribusi atas pendapatan sebesar 51.18 dollar selama 12 bulan siklus hidup mereka.

Secara rata-rata, setiap pelanggan dari kohort kedua berkontribusi atas pendapatan sebesar 41.11 dollar selama 11 bulan siklus hidup mereka.
